## Dividend yield

Dividend yield é o indicador responsável por mostrar como está o pagamento de proventos por parte das empresas. Ele mostra a relação entres os dividendos distribuídos e o preço atual da ação da empresa.

Ou seja, o quanto um ativo pagou em proventos (dividendos e juros sobre capital próprio para as ações e aluguéis para FIIs) nos últimos 12 meses, em relação às suas cotações atuais.

DY (unitário) = Valor de dividendos pagos em determinado período / Preço individual da ação antes dessa distribuição de dividendos. Depois, basta multiplicar esse resultado por 100.

Quais fatores influenciam no dividend yield?
* Preço das ações
* Base de dividendos pagos nos últimos meses
* Pagamento de dividendos extraordinários (acontecimentos extraordinários, ex.: a venda de uma filial de uma empresa)
* Payout

## Dividend Payout

Dividend Payout, poder trazer melhores interpretações dos momentos que as empresas vivem e como elas se comportam em relação ao lucro. Por meio desse indicador, os analistas avaliam se empresa está favorecendo a retenção os lucros para crescer ou a remuneração dos acionistas.

Payout é a porcentagem total do lucro líquido que é distribuída para os acionistas como forma de remuneração por meio de dividendos ou juros sobre o capital próprio.

Para calcular o Dividend Payout, basta dividir o valor distribuído para os acionistas pelo total do lucro líquido (valor após pagar todas as despesas, custos e impostos) gerado pela empresa.

Por isso, é comum nos depararmos com as nomenclaturas "empresas de crescimento" e "empresas de dividendos". Primeiro, avalia-se se a companhia é uma empresa de crescimento ou se já está consolidada no mercado. Então, passa-se à análise do DY, bem como foi o histórico de ambos nos últimos anos para ter uma visão de longo prazo.

## Na prática

**Exemplo:**

A empresa Saturno distribui aos seus acionistas, em forma de dividendos, uma parcela 50% de lucro (PAYOUT). Em determinado período o lucro da empresa foi de 100 mil (LUCRO LÍQUIDO), que deve gerar uma distribuição de 50 mil em dividendos para os acionistas. 

A empresa tem um total 50 mil ações. Assim, distribuindo 50 mil em dividendos, a empresa distribuirá dividendos de 1 real por ação.

No momento atual o preço da ação esta 16 reais. Em regra, é considerado o período de 12 meses nesta análise. Sendo assim, a empresa paga um Dividend Yield de 6,25%.

Resumo dos dados:

* Lucro = 100.000
* Número de ações = 50.000
* Payout = 50%
* Dividendos pago por ação = 1
* Valor atual da ação = 16
* DY = 6,25% 

Utilizando os mesmos dados, se a empresa tem novos projetos de investimento capitais e por isso resolveu diminuir seu payout pela metade a fim de fortalecer seu caixa. Agora o valor do dividendo pago por ação é de 0,50 centavos.

Isso mostra que, a variação do payout afeta o dividend yield, pois existe relação direta entre os dois indicadores.

Ou seja, se um determinado investidor tivesse adquirido 100 ações da empresa Saturno a 16,00 reais cada, teria gasto 1.600,00 reais. Já após a distribuição de dividendos, teria recebido 100,00 reais. Isso representa um lucro de exatamente 6,25% sobre o seu investimento inicial.

### Papel

```python
vale3 = yf.Ticker('VALE3.SA')
```

```python
dados_vale3 = vale3.history(start = '2021-04-23', end = '2022-04-23')
```

```python
dados_vale3.Dividends
```

```python
dividendos = dados_vale3.Dividends
dividendos[dividendos!=0]
```

```python
soma_dividendos = sum(dividendos[dividendos!=0])
soma_dividendos
```

```python
cotacao_final = yf.download('VALE3.SA', start = '2021-04-23')['Adj Close'][-1]
cotacao_final
```

```python
dy = soma_dividendos / cotacao_final
```

```python
dy_vale = dy * 100
dy_vale
```
